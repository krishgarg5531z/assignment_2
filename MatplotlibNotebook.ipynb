```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y,label='2x', color='red', linewidth=3 , marker='.' , markersize=10, markeredgecolor='blue' ,linestyle='-')
plt.title("Basic Graph" , fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'black'})

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.xticks([1,2,3,4])
#plt.xticks(np.arange(1, 4, 1))
plt.legend()
plt.show()
```

### BARCHARTS


```python
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

bars =plt.bar(labels, values)
patterns = ['/', 'x', 'o', '-']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('x')
#bars[1].set_hatch('/')
#bars[2].set_hatch('o')
plt.figure(figsize=(1, 2))

plt.show()
```

Histograms

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall , bins=bins, edgecolor='black', color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Overall Rating')
plt.title('Distribution of FIFA Player Ratings', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'black'})


plt.show()
```

pie charts


```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels =  ['Left', 'Right']
colors = ['#ff9999', '#66b3ff']
plt.pie([left,right] , labels=labels , colors=colors, autopct='%.2f %%')
plt.title('foot prefernece of FIFA players')

```

PIE chart 2


```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight ]
plt.style.use('ggplot')

light =fifa.loc[fifa.Weight < 125].count()[0]
medium_light = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy= fifa.loc[(fifa.Weight >= 200)].count()[0]


weights = [light, medium_light, medium, medium_heavy, heavy]
labels = ['Light (<125 lbs)', 'Medium Light (125-150 lbs)', 'Medium (150-175 lbs)', 'Medium Heavy (175-200 lbs)', 'Heavy (>200 lbs)']
# colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

plt.title("weight distribution of fifa players", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'black'})
explode= (0.4, 0.2, 0, 0, 0.4)  # Explode all slices for emphasis
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, startangle=90 , explode=explode)
```

```python
plt.style.use('default')
plt.figure(figsize=(6, 10))
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
revs=fifa.loc[fifa['Club']== 'New England Revolution']['Overall']
labels= ['FC Barcelona', 'Real Madrid' , 'NER']
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True , medianprops={'linewidth': 2, 'color': 'red'})

for box in boxes['boxes']:
    #Set the edge color
    box.set(color='#4286f4')

    #change fill color
    box.set(facecolor='#abcdef')

plt.title("Professional Soccer team comparision")
plt.show()
```
